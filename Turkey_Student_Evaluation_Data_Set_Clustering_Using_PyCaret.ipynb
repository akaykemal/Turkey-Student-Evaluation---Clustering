```python
#import libraries
import pandas as pd
import numpy as np
from pycaret.clustering import *
```

```python
# load data
df=pd.read_csv("turkiye-student-evaluation_generic.csv")
```

```python
setup(df) # setup the data
```

```python
kmeans=create_model('kmeans',num_clusters=5)
```

```python
plot_model(kmeans,plot='elbow') # use elbow method to see how many clusters, hence we need 5
```

```python
plot_model(kmeans) # 2D Cluster PCA Plot
```

```python
plot_model(kmeans,plot='tsne') # 3D TSNE CLuster Plot
```

Using PyCaret, the elbow method has shown us to use 5 clusters instead of 3
